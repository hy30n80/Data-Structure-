- "rat" is an anagram of "art"

Your function should take two strings as input and return True if the two words are anagrams and False if they are not.

You can assume the following about the input strings:

No punctuation
No numbers
No special characters

An array is a mountain array if and only if:

- `len(array)` >= 3
- There exists some `i` with `0 < i < len(array) - 1` such that:
  - `array[0] < array[1] < ... < array[i - 1] < array[i]`
  - `array[i] > array[i + 1] > ... > array[len(array) - 1]`

image.png

```python
def valid_mountain_array(array):
  """
  checks if the input array is a valid mountain array

  >>> valid_mountain_array([3, 4, 6, 5, 4])
  True
  >>> valid_mountain_array([3, 5, 6, 1, 2, 7, 4, 2, 1])
  False
  >>> valid_mountain_array([3, 4, 5, 6, 7])
  False
  >>> valid_mountain_array([1])
  False
  >>> valid_mountain_array([5, 4, 3, 2, 1])
  False
  """
  # YOUR CODE HERE

  length = len(array)

  if length < 3:
    return False

  if array[0] > array[1]:
    return False

  mode = 0

  for i in range(1, length):

    if (mode == 0) and (array[i] < array[i-1]):
      mode = 1

    elif (mode == 1) and (array[i] > array[i-1]):
      return False

    else:
      continue

  if mode==1:
    return True

  else:
    return False

```
